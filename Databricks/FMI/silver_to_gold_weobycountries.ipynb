```


#silver path
forecastGDPIndicatorsTimesAnnual_Silver = basepath + "silver/fmi/forecastGDPAnnual/"
#gold path
forecastGDPCurrentPrice = basepath + "gold/fmi/forecastGDPCurrentPriceAnual"


```

```

#load silver data
forecastGDP = spark.read.format("delta").load(forecastGDPIndicatorsTimesAnnual_Silver)

display(forecastGDP)
```

```

dfSilver =forecastGDP.withColumn("period",forecastGDP.period.cast("Float"))
```

In this use case we've decided to drop some unnecessary columns such as subject_detail, unitMeasure_code and subject_descriptor

```
dfSilver = forecastGDP.drop("subject_detail", "unitMeasure_code", "subject_descriptor")

dfSilver = dfSilver.where(dfSilver.macroeconomicIndicator == "GDP Current Prices USD")

display(dfSilver)
```

```
#save data to gold table

dfSilver.write.format('delta').mode('overwrite').save(forecastGDPCurrentPrice)
```

```
#load and verify that it's been saved properly
forecastGDP = spark.read.format("delta").load(forecastGDPCurrentPrice)
display(forecastGDP)
```
