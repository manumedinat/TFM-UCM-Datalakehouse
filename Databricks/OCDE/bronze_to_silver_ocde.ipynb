```python
dfSilver_GDP_Constant_Price_Annual = dfSilver_GDP_Constant_Price_Annual.withColumn("macroeconomicIndicator", lit("GDP Constant Prices USD"))
dfSilver_GDP_Current_Price_Annual = dfSilver_GDP_Current_Price_Annual.withColumn("macroeconomicIndicator", lit("GDP Current Prices USD"))
dfSilver_GDP_PerCapita_Constant_Prices_Annual = dfSilver_GDP_PerCapita_Constant_Prices_Annual.withColumn("macroeconomicIndicator", lit("GDP per capita Constant Prices USD"))
dfSilver_GDP_PerCapita_Current_Prices_Annual = dfSilver_GDP_PerCapita_Current_Prices_Annual.withColumn("macroeconomicIndicator", lit("GDP per capita Current Prices USD"))
```

```python
#Joining of the cleaned datasets
df1 = dfSilver_GDP_Constant_Price_Annual.union(dfSilver_GDP_Current_Price_Annual)
df2 = df1.union(dfSilver_GDP_PerCapita_Constant_Prices_Annual)
df_final = df2.union(dfSilver_GDP_PerCapita_Current_Prices_Annual)

df_final.count()
```

```python
display(df_final)
```

```python
#Renaming of columns accordingly to the ICs
df_final2 = df_final.withColumnRenamed("Country","country_desc") \
    .withColumnRenamed("MEASURE","measure") \
    .withColumnRenamed("TIME","period") \
    .withColumnRenamed("Unit","unitMeasure_code") \
    .withColumnRenamed("PowerCode","typeOfData") \
    .withColumnRenamed("Value","value") \
    .withColumnRenamed("Flags","typeOfValue")
```
